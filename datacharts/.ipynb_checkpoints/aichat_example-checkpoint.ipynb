```python
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly

```

```python
# Создаем датафрейм с данными временного ряда
data = pd.read_csv('inoutdata1.csv')
data['ds'] = pd.to_datetime(data['ds'],dayfirst=True)

```

```python
# Создаем и обучаем модель Prophet
model = Prophet()
model.fit(data)
```

```python
# Создаем фрейм для прогнозирования будущих значений
future = model.make_future_dataframe(periods=35)  # Прогноз на 365 дней вперед
```

```python
# Прогнозируем значения
forecast = model.predict(future)

# Выводим результаты прогноза
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
plot_plotly(model, forecast)
```
